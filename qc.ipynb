# Per tile sequence quality issue

Per tile sequence quality was a persistent failure for reverse reads after a slidingwindow 4:15 trim

Before:

<img src="figs/raw_per_tile.png" alt="Before trim" width="400" />

After:

<img src="figs/trimmed_415_per_tile.png" alt="After 4:15 trim" width="400" />

Concerned that the persistent low quality on the same part of reads could lead to false signal, I made the slidingwindow a bit more strict 4:25

4:25 showed only slight improvement, although quite a few more reads were trimmed with the reverse read average sequence length dropping from ~215 to ~200
