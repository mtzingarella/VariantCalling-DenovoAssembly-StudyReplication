```bash
# Count variants by QUAL score bin: <10, 10-30, 30-100, >=100
bcftools view -H ../../results/03_variantcalling/07_genotype_gvcfs/ecoli_8samples.raw.vcf.gz | 
awk '
  BEGIN{w=0; x=0; y=0; z=0}
  {if($6 < 10){w+=1}}
  {if($6 >= 10 && $6 < 30){x+=1}}
  {if($6 >= 30 && $6 < 100){y+=1}}
  {if($6 >= 100){z+=1}}
  END {print "QUAL<10:", w, "\nQUAL 10-30:", x, "\nQUAL 30-100:", y, "\nQUAL>=100:", z}'
```
